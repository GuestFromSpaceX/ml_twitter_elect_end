```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('./new_data/accident.csv')
df
```

```python
text_no_link = []
```

```python
text_column_list = [df['text'][elem] for elem in range(0, 7613)]
print(text_column_list[38])
```

```python
#очистка от http
for elem in text_column_list:
    splited_elem = elem.split(' ')
    for i in range(0, len(splited_elem)):
        if splited_elem[i].startswith('htt') or ('htt' in splited_elem[i]) or ('@' in splited_elem[i]):
            splited_elem[i] = ''
    added_elem = ' '.join(splited_elem)   
    text_no_link.append(added_elem)
text_no_link[38]
```

```python
text_no_link
```

```python
import string as string
```

```python
text_list_no_links_symbols = []
```

```python
#очистка данных от мусора
symbols_to_safe = list(string.ascii_letters)
for elem in text_no_link:
    elem_list = list(elem)
    for i in range(0, len(elem_list)):
        if elem_list[i] not in symbols_to_safe:
            elem_list[i] = ' '
    added_elem = ''.join(elem_list)   
    text_list_no_links_symbols.append(added_elem)
text_list_no_links_symbols[38]
```

```python
text_list_no_links_symbols
```

```python
text_list_no_links_symbols_spaces = []
```

```python
#очистка данных от пробелов
for elem in text_list_no_links_symbols:
    elem_to_words = elem.split(' ')
    temp = []
    for i in range(0, len(elem_to_words)):
        if elem_to_words[i] != '':
            temp.append(elem_to_words[i])
    temp = ' '.join(temp)
    text_list_no_links_symbols_spaces.append(temp)
text_list_no_links_symbols_spaces[38]
```

```python
#очистка от стопслов, некоторых абревиатур и слов меньше 2 символов
for elem in text_2:
    splited_elem = elem.split(' ')
    for i in range(0, len(splited_elem)):
        if (splited_elem[i] in stop_words) or ((len(splited_elem[i]) <= 2) and (splited_elem[i] != 'no')) or (splited_elem[i] in ['amp', 'cdt', 'aba', 'new']): #  
            splited_elem[i] = ''
    added_elem = ' '.join(splited_elem)   
    text_2_no_stopw.append(added_elem)
text_2_no_stopw[38]
```

Теперь нужо проверить, какие слова сколько раз встречаются.

```python
#from nltk.stem import WordNetLemmatizer
#nltk.download('wordnet')
```

```python
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer(language = 'english')
```

```python
#lemmatizer = WordNetLemmatizer()
lemmatizer = stemmer
text_2_no_stopw_lemmatized = []
for elem in text_2_no_stopw:
    # Токенизация предложения на слова
    words = nltk.word_tokenize(elem)
    
    # Лемматизация каждого слова
    lemmatized_words = [lemmatizer.stem(word) for word in words] #stem заменить на lemmatize, если менять
    
    # Соединение лемматизированных слов обратно в предложение
    lemmatized_sentence = " ".join(lemmatized_words)
    
    # Добавление лемматизированного предложения в список
    text_2_no_stopw_lemmatized.append(lemmatized_sentence)
text_2_no_stopw_lemmatized
```

^код на отправку без удаления слов, которые встречаются меньше 2 раз

```python
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
```

```python
all_text_2_no_stopw_lemmatized = ' '.join(text_2_no_stopw_lemmatized)
```

```python
tokens = word_tokenize(all_text_2_no_stopw_lemmatized)
tokens
```

```python
fdist = FreqDist(tokens)
len(fdist)
```

```python
rare_words = []
for word, frequency in fdist.most_common():
    print(f"{word}: {frequency}")
    if frequency <= 0:
        rare_words.append(word)
```

```python
rare_words
```

```python
text_3_no_rare_words = []
```

```python
#очистка от слов, которые встречаются меньше 2 раз
for elem in text_2_no_stopw_lemmatized:
    splited_elem = elem.split(' ')
    for i in range(0, len(splited_elem)):
        if (splited_elem[i] in rare_words):
            splited_elem[i] = ''
    added_elem = ' '.join(splited_elem)   
    text_3_no_rare_words.append(added_elem)
text_3_no_rare_words[38]
```

%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%

```python
all_text_3 = ' '.join(text_3_no_rare_words)
tokens_2 = word_tokenize(all_text_3)
fdist2 = FreqDist(tokens_2)
print(len(fdist2))
```

```python
for word, frequency in fdist2.most_common():
    print(f"{word}: {frequency}")
```
