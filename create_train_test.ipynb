```python
df = pd.read_csv('./new_data/accident_normalized_text.csv')
df
```

```python
duplicates = df['text_normalized'].duplicated()
print(df['text_normalized'][duplicates])
```

```python
df.loc[df.duplicated(subset="text_normalized"), "text_normalized"] = np.nan
df
```

```python
df_no_dup = df.dropna(subset=['text_normalized'], inplace=False)
```

```python
len(df_no_dup)
```

```python
df_nw = df_no_dup[['origin_id','text_normalized', 'target']]
df_nw
```

```python
train_df = df_nw.sample(frac=0.75, random_state=42)  # 75% данных для тренировочного датасета
test_df = df_nw.drop(train_df.index)  # Оставшиеся 25% данных для тестового датасета
```
